```python
from ultralytics import YOLO
import cv2
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# Load model
model = YOLO("yolov8n.pt")

# Load video
cap = cv2.VideoCapture("../data/videoplayback.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Tracking
    results = model.track(frame, persist=True)

    # Draw results
    annotated_frame = results[0].plot()

    # Convert BGR to RGB
    frame_rgb = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)

    # Show in notebook
    plt.imshow(frame_rgb)
    plt.axis('off')
    display(plt.gcf())
    clear_output(wait=True)

cap.release()
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("data/videoplayback.mp4")

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        print("Video Ended")
        break

    # Skip frames (speed control)
    frame_count += 1
    if frame_count % 5 != 0:
        continue

    results = model.track(frame, persist=True)
    annotated_frame = results[0].plot()

    frame_rgb = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)

    plt.imshow(frame_rgb)
    plt.axis('off')
    display(plt.gcf())
    clear_output(wait=True)

    time.sleep(0.3)  # 🔥 IMPORTANT: slow down

cap.release()
```

```python
from ultralytics import YOLO
import cv2

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture("data/videoplayback.mp4")

# Get video properties
width = int(cap.get(3))
height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video save path
out = cv2.VideoWriter(
    "outputs/output.mp4",
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (width, height)
)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Video Ended")
        break

    results = model.track(frame, persist=True)
    annotated_frame = results[0].plot()

    # Save frame
    out.write(annotated_frame)

cap.release()
out.release()

print("✅ Video saved in outputs/output.mp4")
```

```python
from ultralytics import YOLO
import cv2

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture("../data/videoplayback.mp4")

# video properties
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("../outputs/output.mp4", fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    if not ret:
        print("Video Ended")
        break

    results = model.track(frame, persist=True)
    annotated_frame = results[0].plot()

    out.write(annotated_frame)

cap.release()
out.release()

print("✅ Video saved in ../outputs/output.mp4")
```
